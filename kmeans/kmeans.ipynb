```python
from sklearn.datasets import make_blobs 
X, y = make_blobs(n_samples= 1000, centers = 6, n_features = 2, random_state = 0)
print(X.shape)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from kmeans import Kmeans
kmeans = Kmeans(X, 6)
```

```python
kmeans.train()
```

```python
pred = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=pred)
```
